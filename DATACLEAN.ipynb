```python
import pandas as pd
import numpy as np
import datetime
```

```python
df=pd.read_csv("Resources/CAERS_ASCII_2004_2017Q2.csv")
df.tail(1)
```

```python
df=df.drop(columns=['PRI_FDA Industry Code','PRI_Product Role','AEC_Event Start Date'])
df=df.dropna()
df.tail(1)
```

```python
df=df.loc[df["CI_Gender"] != "Not Available"]
df=df[df["CI_Age Unit"]=="Year(s)"]
df=df[df["PRI_FDA Industry Name"]=="Vit/Min/Prot/Unconv Diet(Human/Animal)"]
df=df.drop(columns=['PRI_FDA Industry Name',"CI_Age Unit"])
```

```python
def type(x):
    if "VITAMINS" in x:return("MULTIVITAMIN")
    elif "MULTIVITAMINS" in x:return("MULTIVITAMIN")
    elif "CENTRUM" in x:return("MULTIVITAMIN")
    elif "HYDROXYCUT" in x:return("MULTIVITAMIN")
    elif "VITAMIN C" in x:return("VITAMIN C")
    elif "VITAMIN D" in x:return("VITAMIN D")
    elif "CITRACAL" in x:return("CALCIUM")
    elif "METABO" in x:return("CALCIUM")
    elif "VITAMIN B" in x:return("VITAMIN B")
    elif "SUPER BETA PROSTATE" in x:return("ZINC")
    elif "PRESERVISION" in x:return("VITAMIN D")
    elif "MULTI VITAMIN" in x:return("MULTIVITAMIN")
    elif "FIBER" in x:return("DIETARY FIBER")
    elif "ONE A DAY" in x:return("MULTIVITAMIN")
    elif "MULTI-VITAMIN" in x:return("MULTIVITAMIN")
    elif "B12" in x:return("VITAMIN B")
    elif "OMEGA" in x:return("FISH OIL")
    elif "EMERGEN-C" in x:return("MULTIVITAMIN")
    elif "PROBIOTIC" in x:return("PROBIOTIC")
    elif "OIL" in x:return("FISH OIL")
    elif "ENERGY" in x:return("MULTIVITAMIN")
    elif "VITAMIN E" in x:return("VITAMIN E")
    elif "CALCIUM" in x:return("CALCIUM")
    elif "NIACIN" in x:return("VITAMIN B")
    elif "B-COMPLEX" in x:return("VITAMIN B")
    elif "B COMPLEX" in x:return("VITAMIN B")
    elif "HERBAL" in x:return("MULTIVITAMIN")    
    elif "D3" in x:return("VITAMIN D")
    else:return(x)
    
def prod(x):
    if "CENTRUM" in x:return("CENTRUM")
    elif "HYDROXYCUT" in x:return("HYDROXYCUT")
    elif "PRESERVISION" in x:return("PRESERVISION")
    elif "CITRACAL" in x:return("CITRACAL")
    elif "BENEFIBER" in x:return("BENEFIBER")
    elif "SUPER BETA PROSTATE" in x:return("SUPER BETA PROSTATE")
    elif "ONE A" in x:return("ONE A DAY")
    elif "HERBAL" in x:return("HERBALIFE")
    elif "PROTANDIM" in x:return("PROTANDIM")
    elif "OXYELITE" in x:return("OXYELITE")
   

s0="NON-SERIOUS INJURIES"
s1='VISITED A HEALTH CARE PROVIDER'
s2="HOSPITALIZATION"
s3="LIFE THREATENING"
s4="DEATH"
s5="DISABILITY"

def outcome(x):
    if s4 in x:
        return(s4)
    elif s3 in x:
        return(s3)
    elif s5 in x:
        return(s5)
    elif s2 in x:
        return(s2)
    elif s0 in x:
        return(s0)
    elif s1 in x:
        return(s0)
    else:
        return("IMPORTANT MEDICAL EVENT")
    
def rank(x):
    if x==s4:
        return(6)
    elif x==s3:
        return(5)
    elif x==s5:
        return(4)
    elif x==s2:
        return(3)
    elif x==s0:
        return(1)
    else:
        return(2)
```

```python

df["Type"]=df["PRI_Reported Brand/Product Name"].apply(type)
df["Type"].value_counts()
```

```python
df["Brand/Product"]=df["PRI_Reported Brand/Product Name"].apply(prod)
```

```python
df["Outcome"]=df["AEC_One Row Outcomes"].apply(outcome)
```

```python
df["RA_CAERS Created Date"]=pd.to_datetime(df["RA_CAERS Created Date"])
df["Year"]=df["RA_CAERS Created Date"].dt.year
df["Year/Month"]=df["RA_CAERS Created Date"].dt.to_period('M')
df["Severity"]=df["Outcome"].apply(rank)
```
